```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import Imputer
import pandas as pd
import numpy as np

data = pd.read_csv("sampleXYZ.csv")
data_train = []
target_train = []
data_test = []
target_test = []

data_label = [['X', 'Y', 'Z'], ['X.1', 'Y.1', 'Z.1'], ['X.2', 'Y.2', 'Z.2'], ['X.3', 'Y.3', 'Z.3'], ['X.4', 'Y.4', 'Z.4']]
target_label = ['Activity', 'Activity.1', 'Activity.2', 'Activity.3', 'Activity.4']

additional_labels = np.array([])
for label in data_label:
    for axis in label:
        data[axis+'.average'] = pd.Series(data[axis].rolling(10).mean(), index=data.index)
        additional_labels = np.append(additional_labels, axis+'.average')
for i, label in enumerate(data_label):
    data_label[i] = label + np.reshape(additional_labels, (-1,3)).tolist()[i]
print(data_label)

for label in data_label:
    if len(data_train) == 0:
        data_train = data[label][:3000].values
    else:
        data_train = np.concatenate((data_train, data[label][:3000].values), axis=0)
    if len(data_test) == 0:
        data_test = data[label][3000:4500].values
    else:
        data_test = np.concatenate((data_test, data[label][3000:4500].values), axis=0)

for label in target_label:
    if len(target_train) == 0:
        target_train = data[label][:3000].values
    else:
        target_train = np.concatenate((target_train, data[label][:3000].values), axis=0)
    if len(target_test) == 0:
        target_test = data[label][3000:4500].values
    else:
        target_test = np.concatenate((target_test, data[label][3000:4500].values), axis=0)

#print(len(data_train), len(target_train))


# Get error rate of test data
estimator = Pipeline([("imputer", Imputer(missing_values='NaN',
                                          strategy="mean",
                                          axis=0)),
                      ("svm", SVC(kernel='rbf', C=1.0))])
#estimator = SVC(kernel='rbf', C=1.0)
estimator.fit(data_train, target_train)
label_predict = estimator.predict(data_test)
accuracy = (estimator.score(data_test, target_test))*100
print('Accuracy (test data): ', accuracy, '%')
error_rate = 100 - accuracy
print('Error Rate (test data): ', error_rate, '%')
cm = confusion_matrix(target_test, label_predict)
print('Confusion matrix: \n', cm)

# Get error rate of training data
label_predict = estimator.predict(data_train)
accuracy = (estimator.score(data_train, target_train))*100
print('Accuracy (train data): ', accuracy, '%')
error_rate = 100 - accuracy
print('Error Rate (train data): ', error_rate, '%')
cm = confusion_matrix(target_train, label_predict)
print('Confusion matrix: \n', cm)
```
